## Prepare the query_df

```python
new_cols = query_df.loc[3].values
query_df = query_df.loc[4:]
dictionary = dict(zip(query_df.columns.values, new_cols))
```

```python
query_df
```

```python
columns_to_convert = query_df.columns[1:]  # Skip the first column

for col in columns_to_convert:
    query_df[col] = query_df[col].str.replace(' cm', '').astype(int)

print(query_df)
```

```python
query_df.rename(columns=dictionary, inplace=True)
new_df = query_df.reset_index(drop=True)
```

```python
query_df.columns
```

```python
for index, row in query_df.iterrows():
    new_row = {
        'ProductName': 'Oversized Sweater', # this would be the product name
        'ProductCode': 'OHAPRIL_655bf399580e09abaee75b0f', # this would be the product code
        'Chest': row['Brustweite'],
        'Waist': 0,
        'Hip': row['Saumweite'], # assuming hip matches Saumweite
        'Height': row['Gesamtlänge'], # assuming height matches Gesamtlänge
        'ProductSize': row['Size']
    }
    standard_df = standard_df.append(new_row, ignore_index=True)
```
